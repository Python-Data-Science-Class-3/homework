## Decide about missing values (fill or drop)

```python
# empty str became NAN
df.replace('', np.nan, inplace=True)
df
```

```python
# If there is an Income value, let Employment status be 1
df.loc[(df['Employment Status'].isnull()) & (df['Income'].notnull()), 'Employment Status'] = 1
```

```python
df['Employment Status'].value_counts()
```

```python
# changed the age of 500 to 50.
df.loc[df['Age'] == 500, 'Age'] = 50
```

```python
age_mean = df.loc[df['Age'] >= 18, 'Age'].mean()

# have replaced the values in the 'Age' column less than 18 with the average value.
df.loc[df['Age'] < 18, 'Age'] = age_mean
```

# 2- Categorization

## Categorize satisfaction in 3 level and label them as 'Low', 'Moderate' and 'High'. Add a new column named 'Satisfaction Category' and store them there

#### ||| Low Satisfaction : (1-3) ||| Moderate Satisfaction : (4-7) ||| High Satisfaction : (8-10) |||

```python
def categorize_satisfaction(score):
    if score >= 8:
        return 'High Satisfaction'
    elif score >= 4:
        return 'Moderate Satisfaction'
    else:
        return 'Low Satisfaction'
    
df['Satisfaction Category'] = df['Satisfaction Level'].apply(categorize_satisfaction)
df
```

## Categorize income in 3 level and label them as 'Low', 'Moderate' and 'High'. Add a new column named 'Income Category' and store them there
