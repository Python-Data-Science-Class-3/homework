```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
data_origin = pd.read_csv("global_terror.csv", encoding='latin-1')
data = data_origin.copy()
data.head()
```

-Year       : Indicates the year in which the terrorist attack was made.
-Month      : Indicates the month of the terrorist attack.
-Day        : Indicates the day of the terrorist attack.
-Country    : Indicates in which country the terrorist attack was made.
-State      : Indicates which state the terrorist attack was made.
-Region     : Indicates the region in which the terrorist attack was made.
-City       : Indicates which city the terrorist attack was made in.
-Latitude   : Shows the latitude of the attack.
-Longitude  : Shows the longitude of the attack.
-Attack_type: Shows the way the attack is done.
-Killed     : Determines the number of people killed in the attack.
-Wounded    : Determines the number of people injured in the attack.
-Target     : Indicates what the target of the attack is.
-Group      : Indicates which group the attack was made by.
-Target_type: Which institution or organization (Goverment-Military-Business) is the target of the attack?
-Weapon_type: Which vehicle was used to attack (Explosive-Firearm)
-Motive     : What the attack was for. (Protest-Sabotage)

There are 135 columns in the dataset. We rename those that provide meaningful insights as we analyze.

```python
data.rename(columns = {'iyear':'Year', 'imonth':'Month', 'iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region',
                        'attacktype1_txt':'Attack_type','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary',
                        'gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
df = data[['Year','Month','Day','Country','State','Region','city','latitude','longitude','Attack_type',
           'Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Fill null values with 0 in specific columns
columns = ['State', 'city', 'latitude', 'longitude', 'Killed', 'Wounded', 'Target', 'Summary', 'Motive']
df[columns] = df[columns].fillna(0)
```

```python
# Filter the "Unknown" groups and select the 10 most attacking groups
top_10_groups = df[df['Group'] != 'Unknown']['Group'].value_counts().nlargest(10)

# Draw a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_groups.index, y=top_10_groups.values)
plt.xticks(rotation=45)

# Axis and title settings

plt.xlabel('Group')
plt.ylabel('Number of Attacks')
plt.title('Top 10 Groups (Excluding "Unknown") with the Highest Number of Attacks')

# Show the chart
plt.tight_layout() #automatically adjust the layout of the graphics
plt.show()
```

```python
# Filter the dataframe for entries with "Year" between 2015 and 2017 (inclusive) and "Country" as "Turkey"
filtered_df = df[(df['Year'] >= 2010) & (df['Year'] <= 2017) & (df['Country'] == 'Turkey')]

# Group the data by Year and calculate the sum of Killed
grouped_df = filtered_df.groupby('Year')['Killed'].sum().reset_index()

# Create a bar plot for Year vs. Total Killed
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Killed', data=grouped_df)

# Set axis labels and title
plt.xlabel('Year')
plt.ylabel('Total Number of People Killed')
plt.title('Year vs. Total Number of People Killed in Turkey (2015-2017)')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the dataframe for entries with "Country" as "Turkey" and "Wounded" or "Killed" greater than 5
filtered_df = df[(df['Country'] == 'Turkey') & ((df['Wounded'] > 5) | (df['Killed'] > 10))]

# Create a box plot for Wounded and Killed
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_df[['Wounded', 'Killed']])

# Set axis labels and title
plt.xlabel('Status')
plt.ylabel('Number of People')
plt.title('Number of People Wounded and Killed in Turkey (Wounded or Killed > 5)')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the dataframe for entries where the "Year" column is 2016 and "Weapon_type" is in the specified values
filtered_df = df[(df["Year"] >= 2016) & df["Weapon_type"].isin(["Explosives", "Firearms", "Incendiary", "Melee"])]

# Create a histogram density plot
plt.figure(figsize=(4, 6))
sns.histplot(data=filtered_df, x="Weapon_type", kde=False)

# Set the axis labels and title
plt.xlabel("Weapon Type")
plt.ylabel("Density")
plt.title("Distribution of Weapon Types in 2016 (Explosives, Firearms, Incendiary, Melee)")

# Show the plot
plt.tight_layout()
plt.show()
```
